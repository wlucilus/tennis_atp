# ATP 2024 matches analysis

### This is an analysis of player rankings and wins over the year 2024


### Key results: The players Joao Fonseca, Soon Woo Kwon,  Yi Zhou and Andres Andrade are climbing the rankings fast with Joao And Soon Woo Kwon having exponential trends.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("atp_matches_2024.csv")
df.head(50)
```

```python
for col in df.columns:
    print(col)
```

## Histogram of number of wins

```python
#Count the raw number of wins

#select top 30
df_winners = df.loc[ (df['winner_rank']<30)  ]

wins_count = df_winners['winner_name'].value_counts()
print(wins_count)


#plot in a histogram
wins_count.sort_values(ascending= False).plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Player name')
plt.ylabel('raw Number of wins')
plt.title('Number of wins Top 30')
plt.xticks(rotation=85)  # Rotate labels vertically

# Show the plot
plt.show()

```

```python
#Count the raw number of wins

#select top 100-200
df_winners = df.loc[ (df['winner_rank']>100) &  (df['winner_rank']<200) ]

wins_count = df_winners['winner_name'].value_counts()

#print the 10 biggest winners
print(wins_count.head(20))


#plot in a histogram
wins_count.head(20).sort_values(ascending= False).plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Player name')
plt.ylabel('raw Number of wins')
plt.title('Number of wins Top 100-200')
plt.xticks(rotation=85)  # Rotate labels vertically

# Show the plot
plt.show()


```

## Calculate the win curve of some special player


```python
#some formating of the tournament date , so it becames week of the year.
df['tourney_date'].unique()

# Convert to datetime format
df['date'] = pd.to_datetime(df['tourney_date'].astype(str), format='%Y%m%d')

# Get ISO week number
df['week'] = df['date'].dt.isocalendar().week
#df.head(10)['week']
```

```python
#find id of a specific player
#df.loc[df['winner_name']==pname]['winner_id']
```

```python
#get all players's data

pname = ['Jannik Sinner','Joao Fonseca', 'Grigor Dimitrov', 'Felix Auger Aliassime']

for player in pname:
    df_p_w = df.loc[df['winner_name']==player ].copy()
    df_p_l = df.loc[df['winner_name']==player ].copy()

    #reset the index, so that the cummulative number of wins matches the index number. 
    #first order by tournament data/number, so the cummulative wins are chronological
    df_p_w = df_p_w.sort_values('tourney_date')
    df_p_w.reset_index(drop=True, inplace=True)


    df_p_w['cumulative_wins'] = df_p_w.index

    #plot how many wins he had vs time
    plt.plot( df_p_w['week'], df_p_w['cumulative_wins'], label='Total # of wins')
    plt.plot( df_p_w['week'], df_p_w['winner_rank'], label='Player Ranking')

    #plt.plot( df_p_l['tourney_id'], df_p_l['cumulative_wins'])


    # Add labels and title
    plt.xlabel('Weeks of the year')
    plt.ylabel('')
    plt.title(f'Player:{player}')
    #plt.xticks(rotation=85)  # Rotate labels vertically

    plt.legend()
    plt.show()
```

## Ranking change analysis

```python
# Calculate future prospect, the ones that climbed the ranking faster
# or the ones who have the highest ranking increase through the weeks

player_ids =   df['winner_name'].unique()
ranking_changes = []  # store (player, ranking change)


for player in player_ids:
    
    # calculate initial and final ranking (from wins)
    initial_ranking = df.loc[df['winner_name'] == player].sort_values('week')['winner_rank'].iloc[0]
    final_ranking = df.loc[df['winner_name'] == player].sort_values('week')['winner_rank'].iloc[-1]
    ranking_change =   initial_ranking - final_ranking

    age = df.loc[df['winner_name'] == player]['winner_age'].max()
    ranking_changes.append((player, ranking_change,age ))

ranking_df = pd.DataFrame(ranking_changes, columns=['Player', 'Ranking Change', 'latest Age'])
ranking_df = ranking_df.sort_values('Ranking Change', ascending=False)

print('All players list')
print(ranking_df.head(50))



#select the sub 21 list
sub21_name = ranking_df.loc[ ranking_df['latest Age'] < 22] 
print('sub 21 list')
print(sub21_name.head(20))

#select post 21 list
pos21_name = ranking_df.loc[ ranking_df['latest Age'] > 22] 
print('post 21 list')
print(pos21_name.head(20))
```

```python


pname =  sub21_name.head(15)['Player'].unique()


for player in pname:
    df_p_w = df.loc[df['winner_name']==player ].copy()

    #reset the index, so that the cummulative number of wins matches the index number. 
    #first order by tournament data/number, so the cummulative wins are chronological
    df_p_w = df_p_w.sort_values('tourney_date')
    df_p_w.reset_index(drop=True, inplace=True)


    df_p_w['cumulative_wins'] = df_p_w.index

    delta = ranking_df.loc[ranking_df['Player'] == player]['Ranking Change'].unique()[0]
    age = ranking_df.loc[ranking_df['Player'] == player]['latest Age'].unique()[0]
    plt.plot( df_p_w['week'], df_p_w['winner_rank'], label=f'{player} {age} - delta:{delta}', marker='.')

    #plt.plot( df_p_l['tourney_id'], df_p_l['cumulative_wins'])


    # Add labels and title
    plt.xlabel('Weeks of the year')
    plt.ylabel('Ranking')
    plt.title(f'biggest ranking trends under 22yrs')
    #plt.xticks(rotation=85)  # Rotate labels vertically

plt.legend(loc = 'center left', bbox_to_anchor=(1,0.5))
plt.show()
```

### Is 1st percentage % correlated to ranking change?
### and 2nd serve win %?


```python
for player in ['Joao Fonseca', 'Yi Zhou']:
    df_p_w = df.loc[df['winner_name']==player ].copy()
    df_p_w.sort_values('week', inplace=True)
    
    #plot how many wins he had vs time
    plt.plot( df_p_w['week'], df_p_w['w_1stIn'], label=f'{player} - 1st serve in', marker='+')
    plt.plot( df_p_w['week'], df_p_w['w_1stWon'], label=f'{player} - 1st serve won', marker='+')
    plt.plot( df_p_w['week'], df_p_w['w_2ndWon'], label=f'{player} - 2nd serve won', marker='+')

    
    # Add labels and title
    plt.xlabel('Weeks of the year')
    plt.ylabel('% percentages')
    plt.title(f'biggest ranking trends')
    #plt.xticks(rotation=85)  # Rotate labels vertically

plt.legend(loc = 'center left', bbox_to_anchor=(1,0.5))
plt.show()
```
